## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

# 1. Dropping unnecessary columns
df.drop(columns = 'Unnamed: 0', inplace= True)

# 2. Salary Estimate-
# Filtering out records with salary < 0
df = df[df['Salary Estimate']!= '-1']

# Splitting the Salary Estimate Column and view only the salary ranges
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])

# Removing the $ and K symbols from salary estimate column to view only salary range values
df['Salary Estimate']= df['Salary Estimate'].apply(lambda x: x.replace('$','').replace('K',''))

# Creating new column which stores all the salaries paid on hourly basis
df['Hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)

# Creating new column which stores the records of all the employer provided salaries
df['Employer_Provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)

# Removing the text i.e per hour, employer provided salary containing in salary estimate column
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))

# Extracting minimum and maximum salary
df['Min_Salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[0]))
df['Max_Salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[1]))
# Calculating average salary from minimum and maximum values
df['Average_Salary'] = (df['Min_Salary'] + df['Max_Salary'])/2

# Ensuring salaries are in full values
df['Min_Salary'] = df['Min_Salary'] * 1000
df['Max_Salary'] = df['Max_Salary'] * 1000

# Converting all the salaries which are hourly basis into annual basis
df['Min_Salary'] = df.apply(lambda x: x.Min_Salary*40*52 if x['Hourly']== 1 else x.Min_Salary, axis = 1)
df['Max_Salary'] = df.apply(lambda x: x.Max_Salary*40*52 if x['Hourly']== 1 else x.Max_Salary, axis = 1)

# 3. Company Name-
# Extracting only Company Name part
df['Company_Text'] = df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-4], axis=1)

# 4. Location
# Finding State where company has physical existence
df['Job_State'] = df['Location'].apply(lambda x: x.split(',')[1])    # Splitting the values in location and extracting the state part

# Finding Records which have job location is in headquarters
df['Same_State'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis = 1)

# 5. Founded-
# Finding the company age from current year (number of years company existing)
df['Company_Age'] = df['Founded'].apply(lambda x: x if x < 1 else 2025-x)

# 6. Job Description-
# Extracting skills from job description
skills = ['Python', 'SQL', 'Tableau', 'Excel', 'Machine Learning', 'Deep Learning', 'Power BI', 'R Studio', 'AWS']
# Create a new column for each skill
for skill in skills:
    df[skill] = df['Job Description'].str.contains(skill, case=False, na=False).astype(int)

# 7. Job Title-
# Extracting consistent job title
def title_simplifier(title):
  if 'data scientist' in title.lower():
    return 'Data Scientist'
  elif 'data engineer' in title.lower():
    return 'Data Engineer'
  elif 'analyst' in title.lower():
    return 'Analyst'
  elif 'machine learning' in title.lower():
    return 'Machine Learning'
  elif 'manager' in title.lower():
    return 'Manager'
  elif 'director' in title.lower():
    return 'Director'
  else:
    return 'Others'
# Applying the function
df['Job_Simp'] = df['Job Title'].apply(title_simplifier)

# Extracting seniority level from job title
def seniority_level(title):
  if 'sr' in title.lower() or 'senior' in title.lower() or 'sr.' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
    return 'Senior'
  elif 'jr' in title.lower() or 'jr.' in title.lower():
    return 'Junior'
  else:
    return 'NA'
# Applying the function
df['Seniority'] = df['Job Title'].apply(seniority_level)

# 8. Competitors-
# Calculating number of competitors for each company
df['Num_Competitors'] = df['Competitors'].apply(lambda x: len(x.split(',') if x!= -1 else 0))

# View all the columns of the dataset
pd.set_option('display.max_columns', None)
# View top 5 records of the dataset
df.head(5)
```
